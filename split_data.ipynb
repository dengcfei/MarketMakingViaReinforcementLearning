```python
import pandas as pd
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np

# Sample data (replace with your own dataframe)
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=100),
    'Close': np.arange(1,101,1)
})

lookback = 10
target_day = 3
batch_size = 32

# Convert 'Close' column to numpy array
data_array = data['Close'].values

# Adjust the length parameter to include the lookback period and the target day
length = lookback + target_day

# Generate input-output pairs using TimeseriesGenerator
generator = TimeseriesGenerator(data_array, data_array, length=length, batch_size=batch_size)


    
# Print the generated input-output pairs
for i in range(len(generator)):
    inputs, targets = generator[i]
    print("Input sequence:", inputs)
    print("Target value:", targets)  # Print the last value of each sequence (T+3)
    print("---")
```
